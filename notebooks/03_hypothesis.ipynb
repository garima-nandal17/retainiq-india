## Drivers ranked by |effect size|

```python
pd.DataFrame(out['tests']).assign(abs_effect=lambda d: d.effect_size.abs())\
  .sort_values('abs_effect', ascending=False)[['factor','test','effect','effect_size']]
```

**Verdict.** Tenure (Cohen's d ≈ −0.85) and contract type (Cramér's V ≈ 0.41) are the load-bearing drivers; senior-citizen status is statistically real but small (V ≈ 0.15). Cramér's V and Cohen's d are on different scales, so the ranking is a heuristic, not an exact ordering.
